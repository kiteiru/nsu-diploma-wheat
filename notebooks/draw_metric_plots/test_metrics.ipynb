```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = [0.925, 0.929, 0.889, 0.894, 0.897, 0.974, 0.978]
r = [0.874, 0.866, 0.975, 0.972, 0.975, 0.975, 0.968]
f = [0.898, 0.896, 0.925, 0.927, 0.931, 0.973, 0.971]

names = ['Precision', 'Recall', 'F-score']
colors = ['darkviolet', 'orchid', 'royalblue']
model_names = ['Boosted-Unet + MSE', 'Unet + WHD', 'Yolo-Nano', 'Yolo-Middle', 'Yolo-Large', 'Circles + Unet + BCE', 'Ellipses + Unet + BCE']
```

```python
clsfrs = [p, r, f]

x = [i for i in range(7)]

fig = plt.figure(figsize=(8,5))
for i in range(len(clsfrs)):
    y = clsfrs[i]
    plt.plot(x, y, linestyle='--', marker='o', color=colors[i], label=names[i])
    
plt.legend(loc="upper left", bbox_to_anchor=(0.7, 0.32), fontsize=14)

plt.xticks(np.linspace(0, 3, 4))
plt.xticks(x, model_names, rotation=20)
ax = plt.gca()
ax.tick_params(axis='both', labelsize=13)
plt.show()
```

```python
fig.savefig("test_plot.png", bbox_inches='tight')
```
