```python
import cv2
import numpy as np
import pandas as pd
import sys
import os
import csv 

from exif import Image
from read_roi import read_roi_file
from matplotlib import pyplot as plt
from pathlib import Path
from tqdm.notebook import tqdm
#../november-train/
IMG_DIR_PATH = "../november-train/orig_images"
ROI_DIR_PATH = "../november-train/orig_rois"
FEATURES_FILE_PATH = "raw_features.csv"
#%%
table_columns = ['make', 'model', 'x_resolution', 'y_resolution', 
                 'resolution_unit', 'software', 'datetime', 
                 '_exif_ifd_pointer', 'compression', 
                 'jpeg_interchange_format', 'jpeg_interchange_format_length', 
                 'exposure_time', 'f_number', 'exposure_program', 
                 'photographic_sensitivity', 'exif_version', 
                 'datetime_original', 'datetime_digitized', 'shutter_speed_value', 
                 'aperture_value', 'exposure_bias_value', 'subject_distance', 
                 'metering_mode', 'focal_length', 
                 'pixel_x_dimension', 'pixel_y_dimension', 
                 'focal_plane_resolution_unit', 'exposure_mode', 'white_balance', 
                 'spikelets_num', 'name']
                 

#%%
def save_to_file(table):
    table.to_csv(FEATURES_FILE_PATH, index=False)
    
def get_exif_meta(name, ext, row):
    exif_info = Image(IMG_DIR_PATH + "/" + name + ext)
    #all_f = exif_info.list_all()
    #print(exif_info.list_all())
 
    for i in range(len(table_columns) - 2):
        feat = table_columns[i]
        value = exif_info.get(feat)
        row.append(value)
    return row


def main():
    spikelets = []
    
    table = pd.DataFrame(columns=table_columns)
    #print(table)
    
    for file in tqdm([*(Path(IMG_DIR_PATH)).glob('*.jpg')]):
        name, ext = os.path.splitext(os.path.basename(file))
     
        img = cv2.imread(IMG_DIR_PATH + "/" + name + ext)
        roi = pd.read_csv(ROI_DIR_PATH + "/" + name + ".roi2", delimiter=";", header=None)
        
        dots = zip(roi[0], roi[1])
        
        row = []

        row = get_exif_meta(name, ext, row)
        
        spikelets_num = len(roi[0])
        row.append(spikelets_num)
        row.append(name)
        
        table.loc[len(table)] = row

        
    print(table)
    #save_to_file(table)
    table.to_csv(FEATURES_FILE_PATH, index=False, quoting=csv.QUOTE_NONE, escapechar='|')
    
if __name__ == "__main__":
    main()
```
